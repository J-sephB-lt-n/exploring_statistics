# Causal DAGs

for causal theory, please refer to this incredible, incredible book: [The Causal Inference Book](https://www.hsph.harvard.edu/miguel-hernan/causal-inference-book/)

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import seaborn as sns
sns.set_theme()
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
import itertools
```

```python
G = nx.DiGraph()
G.add_edge('\ncause','\neffect',color='green',weight=2)

f,ax = plt.subplots(1,1, figsize=(5,5))

edge_colours = nx.get_edge_attributes(G,'color').values()
edge_weights = nx.get_edge_attributes(G,'weight').values()
node_positions = nx.circular_layout(G) 

plt.title("Building Block of Directed Acyclic Graph")
plt.axis('off')

nx.draw(
        G,  
        node_positions,
        edge_color = edge_colours, 
        width = list(edge_weights),
        with_labels = True,
        node_color = 'lightgrey'
)
```

```python
# Causal Chain
G = nx.DiGraph()
G.add_edge(1,2,color='green',weight=2)
G.add_edge(2,3,color='green',weight=2)

G.add_edge(4,5,color='red',weight=2)
G.add_edge(5,6,color='red',weight=2)

f,ax = plt.subplots(1,1, figsize=(5,5))

edge_colours = nx.get_edge_attributes(G,'color').values()
edge_weights = nx.get_edge_attributes(G,'weight').values()
node_positions = nx.circular_layout(G) 

plt.title("Causal Chain")
plt.axis('off')

nx.draw(
        G,  
        node_positions,
        edge_color = edge_colours, 
        width = list(edge_weights),
        with_labels = True,
        node_color = 'lightgrey'
)

ax.add_patch( Circle(node_positions[5],radius=0.085,linewidth=2,edgecolor='red',facecolor='none'))
```

```python
# Common Effect (Collider)
G = nx.DiGraph()
G.add_edge(1,2,color='red',weight=2)
G.add_edge(3,2,color='red',weight=2)

G.add_edge(4,5,color='green',weight=2)
G.add_edge(6,5,color='green',weight=2)

f,ax = plt.subplots(1,1, figsize=(5,5))

edge_colours = nx.get_edge_attributes(G,'color').values()
edge_weights = nx.get_edge_attributes(G,'weight').values()
node_positions = nx.circular_layout(G) 

plt.title("Common Effect (Collider)")
plt.axis('off')

nx.draw(
        G,  
        node_positions,
        edge_color = edge_colours, 
        width = list(edge_weights),
        with_labels = True,
        node_color = 'lightgrey'
)

ax.add_patch( Circle(node_positions[5],radius=0.085,linewidth=2,edgecolor='green',facecolor='none'))
```

```python
# Common Cause
G = nx.DiGraph()
G.add_edge(1,2,color='green',weight=2)
G.add_edge(1,3,color='green',weight=2)

G.add_edge(4,5,color='red',weight=2)
G.add_edge(4,6,color='red',weight=2)

f,ax = plt.subplots(1,1, figsize=(5,5))

edge_colours = nx.get_edge_attributes(G,'color').values()
edge_weights = nx.get_edge_attributes(G,'weight').values()
node_positions = nx.circular_layout(G) 

plt.title("Common Cause")
plt.axis('off')

nx.draw(
        G,  
        node_positions,
        edge_color = edge_colours, 
        width = list(edge_weights),
        with_labels = True,
        node_color = 'lightgrey'
)

ax.add_patch( Circle(node_positions[4],radius=0.085,linewidth=2,edgecolor='red',facecolor='none'))
```

# Example 1

```python
G = nx.DiGraph()
G.add_edge('customer\nexposure\nto\nadvertising','customer\nbuys\nproduct',color='black',weight=2)
G.add_edge('customer\nmood','customer\nbuys\nproduct',color='black',weight=2)
G.add_edge('customer\nage','customer\nlikes\nproduct',color='black',weight=2)
G.add_edge('customer\nage','customer\namt\nof\nmoney',color='black',weight=2)
G.add_edge('customer\namt\nof\nmoney','customer\nbuys\nproduct',color='black',weight=2)
G.add_edge('customer\nlikes\nproduct','customer\nbuys\nproduct',color='black',weight=2)
G.add_edge('customer\nlikes\ntechnology','customer\nexposure\nto\nadvertising',color='black',weight=2)
G.add_edge('customer\nbuys\nproduct','business\nrevenue',color='black',weight=2)

f,ax = plt.subplots(1,1, figsize=(15,10))

edge_colours = nx.get_edge_attributes(G,'color').values()
edge_weights = nx.get_edge_attributes(G,'weight').values()
node_positions = nx.circular_layout(G) 

plt.title("joe u can label the effects on the edges!")
plt.axis('off')

nx.draw(
        G,  
        node_positions,
        edge_color = edge_colours, 
        width = list(edge_weights),
        with_labels = True,
        node_color = 'lightgrey',
        font_color = 'brown',
        node_size = 3000
)
```

```python
# code to simulate data from this system:
# each variable has probabilty 0.3 of having value 1
# variables cause their childrens prob of 1 to increase/decrease if they have the value 1 
def sim_data_row():
    customer_age = np.random.binomial(n=1, size=1, p=0.3)[0]
    customer_likes_product = np.random.binomial(n=1, size=1, p=0.3 + 0.2*customer_age)[0]
    customer_amount_of_money = np.random.binomial(n=1, size=1, p=0.3 + 0.2*customer_age)[0]
    customer_likes_technology = np.random.binomial(n=1, size=1, p=0.2)[0]
    customer_exposure_to_advertising = np.random.binomial(n=1, size=1, p=0.3 + 0.2*customer_likes_technology)[0]
    customer_mood = np.random.binomial(n=1, size=1, p=0.2)[0]
    customer_buys_product = np.random.binomial(
        n=1, 
        size=1, 
        p = 0.2 + customer_mood*0.1 + customer_likes_product*0.1 + customer_amount_of_money*0.1 + customer_exposure_to_advertising*0.1
    )[0]
    business_revenue = np.random.binomial(n=1, size=1, p=0.2 + 0.5*customer_buys_product)[0]

    return pd.DataFrame(
        {
            'customer_age':[customer_age],
            'customer_likes_product':[customer_likes_product],
            'customer_amount_of_money':[customer_amount_of_money],
            'customer_likes_technology':[customer_likes_technology],
            'customer_exposure_to_advertising':[customer_exposure_to_advertising],
            'customer_mood':[customer_mood],
            'customer_buys_product':[customer_buys_product],
            'business_revenue':[business_revenue]
        }
    )
    
population_size = 10_000
simdata = pd.concat( [sim_data_row() for i in range(population_size)] )
```

```python
vars_in_data = list( simdata.columns )
print( vars_in_data )
all_var_combinations = list( itertools.combinations(vars_in_data, 2) )
store_chisq_results = []

combination_i = 0
for combination_i in range(len(all_var_combinations)):
    var1_name = all_var_combinations[combination_i][0]
    var2_name = all_var_combinations[combination_i][1]
    #print( var1_name, '\t', var2_name )
    g, p, dof, expctd = chi2_contingency( 
                                        pd.crosstab( simdata[var1_name], simdata[var2_name] ).to_numpy(), 
                                        correction = False
                                    )
    #print(p, '\n')
    store_chisq_results.append( 
        pd.DataFrame(
            {
                'var1':[var1_name],
                'var2':[var2_name],
                'p':[p]
            }
        )
    )
    
chisq_results_df = pd.concat( store_chisq_results, axis=0 )
chisq_results_df

```
